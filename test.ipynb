```python
import os
import pandas as pd
from gtts import gTTS

# Load your CSV file
df = pd.read_csv("Words.csv")

# Create folder for pronunciations if it doesn't exist
os.makedirs("Pronunciation", exist_ok=True)

# Loop through the words
for word in df["Word"]:
    if pd.isna(word):  # skip empty cells
        continue
    word_clean = str(word).strip()
    if word_clean == "":
        continue

    try:
        # Generate pronunciation
        tts = gTTS(word_clean, lang="en")
        
        # Save as MP3 inside the folder
        filename = f"www/Pronunciation/{word_clean}.mp3"
        tts.save(filename)
        print(f"Saved: {filename}")
    except Exception as e:
        print(f"Error with word '{word_clean}': {e}")
```
